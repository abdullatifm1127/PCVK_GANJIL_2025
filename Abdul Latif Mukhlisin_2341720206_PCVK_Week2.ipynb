```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("KTM Abdul Latif.jpg")
plt.imshow(img)

img2 =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)

cv2.rectangle(img2, pt1=(50,40), pt2=(1000,290), color=(0,255,255), thickness=20)
plt.imshow(img2)

cv2.rectangle(img2, pt1=(2100,40), pt2=(2750,290), color=(255,255,0), thickness=20)
plt.imshow(img2)

cv2.rectangle(img2, pt1=(600,1620), pt2=(1800,1720), color=(255,0,255), thickness=20)
plt.imshow(img2)
```
