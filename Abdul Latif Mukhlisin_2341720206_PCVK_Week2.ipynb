```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("KTM Abdul Latif.jpg")

img2 =cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Konversi BGR ke RGB
plt.imshow(img2)

# Logo Polinema
cv2.rectangle(img2, pt1=(30,10), pt2=(100,200), color=(0,255,255), thickness=20)
plt.imshow(img2)

# Logo BNI
cv2.rectangle(img2,  pt1=(1180,40), pt2=(1550,170),color=(255,255,0), thickness=20)
plt.imshow(img2)

# Program Studi
cv2.rectangle(img2, pt1=(360,920), pt2=(940,1000), color=(255,0,255), thickness=25)
plt.imshow(img2)
```
